```python
import pandas as pd
```

**Reading of dataset into a dataframe**

```python
df = pd.read_excel('bank-full-dataset.xlsx', engine='openpyxl', header=None)
df.head()
```

**Splitting of data into multiple columns & removal of semi-colon marks**

```python
df_split = df[0].str.split(';', expand=True)
df_split.head()
```

**Removal of characters (*quotation marks*) from values in the dataset**

```python
df_split = df_split.applymap(lambda x: x.strip('"'))
df_split.head()
```

**Putting the first row as the header**

```python
df_split.columns = df_split.iloc[0]
df_split.head()
```

**Removal of first row from the dataset**

```python
df_split = df_split.loc[1:]
df_split.head()
```

**Information on the dataset**

```python
df_split.info()
```

**Summary**

The dataset is collected from UCI Machine Learning Repository in Comma Separated Value (CSV) format. It contains rows of **Customers Information** which are stored in a single column. The main issues  is that the data is not parsed into separate columns and the values in it has some characters such as semicolons and quotation marks these make it inappropriate for analysis. Therefore, the above data cleaning processes centered on separating the dataset into multiple columns and these characters are removed in order to have an analytic dataset.
