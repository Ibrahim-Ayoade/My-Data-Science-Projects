```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
```

# Write a wrangle function to process data

```python
def wrangle (filename):
    # Reading data into filename
    df = pd.read_csv(filename)
    # Subsetting 'Nigeria' from the entire dataset
    df_nigeria = df[df["country"] == "Nigeria"]
    # Removing empty columns from the dataset
    threshold = 1
    df_nigeria = df_nigeria.loc[:, df_nigeria.isna().mean()<threshold]
   
    return df_nigeria
```

```python
filename = "World Energy Consumption.csv"
df = wrangle(filename)
df.head()
```
